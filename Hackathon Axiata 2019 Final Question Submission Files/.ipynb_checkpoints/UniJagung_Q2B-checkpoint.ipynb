## Import libraries and data

```python
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.metrics import mean_squared_error

%matplotlib inline


df = pd.read_json('Axiata Urbanlytics/Dataset/air_sensor.json')
df = df.pivot_table(index = "CreatedDateTime", columns = "Name", values="Value", aggfunc="last")
```

## Create the date and hour attributes

```python
df.index = pd.to_datetime(df.index, unit='s')
df['date'] = [d.date() for d in df.index]
df['hour'] = [d.hour for d in df.index]

data = df[['date', 'hour', 'pm25', 'heatIndex', 'temperature']]
data = data.groupby(['date', 'hour']).mean()

data = data.sort_index()
cols = data.columns
```

## Normalize time series data

```python
from sklearn.preprocessing import MinMaxScaler
```

## prepare data for normalization

```python
scaler = MinMaxScaler()
```

## Creating the train and validation set

```python
train = data[:int(0.8*(len(data)))]
valid = data[int(0.8*(len(data))):]
```

```python
#fit the model
```

```python
from statsmodels.tsa.vector_ar.var_model import VAR

train_normalized = scaler.fit_transform(train)
train.plot(figsize=(15,3))

model = VAR(endog=train_normalized)
model_fit = model.fit(56)
```

## make prediction on validation

```python
prediction = model_fit.forecast(model_fit.y, steps=len(valid))
prediction = scaler.inverse_transform(prediction)
```

## converting predictions to dataframe

```python
pred = pd.DataFrame(prediction, columns=cols)
```

## check rmse

```python
print('rmse value is : ', sqrt(mean_squared_error(pred.pm25, valid.pm25)))

fig, ax = plt.subplots(1, 1, dpi=150, figsize=(10,3))
pred.pm25.plot(legend=True, ax=ax, label='Forecast')
valid.pm25.plot(legend=True, ax=ax, label='Actual')
ax.set_title("PM25: Forecast vs Actuals")
ax.spines["top"].set_alpha(0)
ax.tick_params(labelsize=6)

plt.tight_layout();
```

## make final predictions

```python
data_normalized = scaler.fit_transform(data)

model = VAR(endog=data_normalized)
model_fit = model.fit(56)

# predict 7*24 = 168 steps
actual_pred = model_fit.forecast(model_fit.y, steps=168)
actual_pred = scaler.inverse_transform(actual_pred)

actual_pred = pd.DataFrame(actual_pred, columns=cols)
print(actual_pred.pm25)

actual_pred.pm25.to_csv('q2b.csv', index=False)
```
